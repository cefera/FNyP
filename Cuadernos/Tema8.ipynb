```python
import numpy as np
import matplotlib.pyplot as plt
import pdg

###########################################################
#   JPAC color style
###########################################################

jpac_blue   = "#1F77B4"; jpac_red    = "#D61D28";
jpac_green  = "#2CA02C"; jpac_orange = "#FF7F0E";
jpac_purple = "#9467BD"; jpac_brown  = "#8C564B";
jpac_pink   = "#E377C2"; jpac_gold   = "#BCBD22";
jpac_aqua   = "#17BECF"; jpac_grey   = "#7F7F7F";

jpac_color = [jpac_blue, jpac_red, jpac_green, 
              jpac_orange, jpac_purple, jpac_brown,
              jpac_pink, jpac_gold, jpac_aqua, jpac_grey, 'black' ];

jpac_axes = jpac_color[10]

api = pdg.connect(pedantic=True)

print('PDG edition:',api.edition)
```

```python
def plotbox(x,y,textstr,colorfondo):
    props = dict(boxstyle='round', facecolor=colorfondo, alpha=0.5)
    plt.text(x, y, textstr, transform=ax.transAxes, fontsize=14,
        verticalalignment='center', bbox=props)
    return

fig, ax = plt.subplots(figsize=(5,5))
plotbox(0.05,0.8,r'       fermiones        ',jpac_grey)
plotbox(0.05,0.3,r'        bosones         ',jpac_grey)
plotbox(0.05,0.7,r'quarks',jpac_red)
plotbox(0.32,0.7,r'leptones',jpac_blue)
plotbox(0.07,0.6,r'$u$',jpac_red)
plotbox(0.07,0.5,r'$c$',jpac_red)
plotbox(0.07,0.4,r'$b $',jpac_red)
plotbox(0.17,0.6,r'$d$',jpac_red)
plotbox(0.17,0.5,r'$s $',jpac_red)
plotbox(0.17,0.4,r'$t $',jpac_red)
plotbox(0.36,0.6,r'$e$',jpac_blue)
plotbox(0.36,0.5,r'$\mu $',jpac_blue)
plotbox(0.36,0.4,r'$\tau $',jpac_blue)
plotbox(0.45,0.6,r'$\nu_e$',jpac_blue)
plotbox(0.45,0.5,r'$\nu_\mu $',jpac_blue)
plotbox(0.45,0.4,r'$\nu_\tau $',jpac_blue)
plotbox(0.05,.2,r'mediadores',jpac_purple)
plotbox(0.393,.2,r'masa',jpac_green)
plotbox(0.05,.1,r'$\gamma$',jpac_gold)
plotbox(0.13,.1,r'$g$',jpac_aqua)
plotbox(0.21,.1,r'$Z$',jpac_orange)
plotbox(0.29,.1,r'$W$',jpac_orange)
plotbox(0.45,.1,r'$H$',jpac_green)
ax.set_axis_off()
plt.show()
#fig.savefig("particulas.pdf", bbox_inches='tight')
#fig.savefig("particulas.png", bbox_inches='tight')
```

Ejemplo de uso de la API del PDG. Propiedades del protón.

```python
for p in api.get_particle_by_name('p').properties():
    print('%-60s    %s' % (p.description, p.display_value_text))
```

Ejemplo de uso de la API del PDG. Propiedades del fotón.

```python
for p in api.get_particle_by_name('gamma').properties():
    print('%16s:  %-60s    %s' % (p.pdgid, p.description, p.display_value_text))
```

Ejemplo de uso de la API del PDG. Anchuras de línea porcentuales inclusivas del B+.

```python
for bf in api.get_particle_by_name('B+').inclusive_branching_fractions():
    print('%-80s    %4s    %s' % (bf.description, bf.is_limit, bf.value))
```

Ejemplo de uso de la API del PDG. Anchuras de línea porcentuales exclusivas del B+

```python
for bf in api.get_particle_by_name('B+').exclusive_branching_fractions():
    print('%-80s    %4s    %s' % (bf.description, bf.is_limit, bf.value))
```
