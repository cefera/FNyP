```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import matplotlib.cm as cm
import pandas as pd
import matplotlib.ticker as ticker

jpac_blue   = "#1F77B4"; jpac_red    = "#D61D28";
jpac_green  = "#2CA02C"; jpac_orange = "#FF7F0E";
jpac_purple = "#9467BD"; jpac_brown  = "#8C564B";
jpac_pink   = "#E377C2"; jpac_gold   = "#BCBD22";
jpac_aqua   = "#17BECF"; jpac_grey   = "#7F7F7F";

jpac_color = [jpac_blue, jpac_red, jpac_green, 
              jpac_orange, jpac_purple, jpac_brown,
              jpac_pink, jpac_gold, jpac_aqua, jpac_grey, 'black' ];

jpac_axes = jpac_color[10]
dashes = 60*'-'

# the service URL
livechart = "https://nds.iaea.org/relnsd/v1/data?"

# There have been cases in which the service returns an HTTP Error 403: Forbidden
# use this workaround
import urllib.request
def lc_pd_dataframe(url):
    req = urllib.request.Request(url)
    req.add_header('User-Agent', 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:77.0) Gecko/20100101 Firefox/77.0')
    return pd.read_csv(urllib.request.urlopen(req))

df = lc_pd_dataframe(livechart + "fields=ground_states&nuclides=all")
df = df[pd.to_numeric(df['binding'],errors='coerce').notna()]
df['binding'] /= 1000
df['A'] = df['n'] + df['z']

uma = 931.4941024/1000000/1000

# A: Número de nucleones
# r0: constante, ha de ser 1.22
# a: espesor de la corteza

def radio(r0, A):
    return r0 * (A**(1/3))

def densidadnuclear(r0, A, a, r):
    R = radio(r0, A)
    return 1. / (1. + np.exp((r - R) / a))

def WoodsSaxon(r0, A, a, r):
    R = radio(r0, A)
    return -1. / (1. + np.exp((r - R) / a))
```

Potencial atómico

```python
def coulomb(Z, r):
    return -2.*Z / r

def barrera(l, r):
    return l * (l + 1) / (r * r)

def potencial(Z, l, r):
    return coulomb(Z, r) + barrera(l, r)

def En(ry, nr, l):
    n = nr + l + 1
    return -ry / (n**2)

a0, ry  = 0.529177210903, 13.605693122994
Z, lmax, nmax = 1, 2, 5

fig = plt.figure(figsize=(5,5))
rmin, rmax, rstep = 0.1, 15, 0.01
r = np.arange(rmin, rmax, rstep)
plt.xlim((rmin, rmax))
plt.ylim((-20., 5))
plt.ylabel(r'$V$ (eV)', fontsize=15)
plt.xlabel(r'$r$ ($a_0$)', fontsize=15)
plt.hlines(0, rmin, rmax, colors=jpac_color[9], lw=1., linestyles='solid', alpha=0.5)
for l in range(lmax+1):
    texto = '$\ell=$' +  str(l)
    plt.plot(r, potencial(Z, l, r)*ry,'-', c=jpac_color[l], label=texto)
    plt.hlines(En(ry, 0, l), rmin, rmax, colors=jpac_color[l], lw=1., linestyles='dashed')
    
    for n in range(1, nmax+1):
        plt.hlines(En(ry, n, l), rmin, rmax, colors=jpac_color[l], lw=1., linestyles='dotted', alpha=0.5)

plt.legend(loc='lower right', ncol=1, frameon=True)
plt.show()
#fig.savefig("hidrogeno.pdf", bbox_inches='tight')
```

Radio nuclear vs radio de carga

```python
dfres = df.loc[(df['half_life'] == 'STABLE')].dropna(subset=['A', 'radius'])
dfrne = df.loc[(df['half_life'] != 'STABLE')].dropna(subset=['A', 'radius'])

file_BW = 'BlattWeisskopf.csv'
dfbw = pd.read_csv(file_BW, sep=',', header=0, skiprows=0)

# Singh, S.; Kumar, S.; Singh, B. y Jain, A. K., Nuclear radius parameters (r0) for even-even nuclei from alpha decay. Nucl. Data Sheets 167 (2020) 1
file_NDS = 'NDS2020_167.csv'
dfnds = pd.read_csv(file_NDS,sep=',',header=0,skiprows=1)
dfnds['radio'] = radio(dfnds['r0'],dfnds['AD'])

# I. Angeli and K.P. Marinova, Atomic Data and Nuclear Data Tables 99 (2013) 69-95 https://doi.org/10.1016/j.adt.2011.12.006
def radii(r0,r1,r2,A):
    return A**(1/3)*(r0+ r1/(A**(2/3)) + r2/(A**(4/3)) )

rho0 = 0.15 # Horowitz, Piekarewicz, Reed, Phys. Rev. C 102 (2020) 044321 
radiosaturacion = (3/(4*np.pi*rho0))**(1/3)

fig = plt.figure(figsize=(8,5))
plt.xlim((0.5,6.5))
plt.ylim((0,12))
plt.xlabel(r'$A^{1/3}$ ',fontsize=15)
plt.ylabel(r'Radio (fm)' ,fontsize=15)
x = np.linspace(0.1,6.5,100)
plt.plot(x,1.23*x,'--',c=jpac_color[1],lw=1,label=r'$R=1,22\, A^{1/3}$')
plt.plot(x,1.5*x,'--',c=jpac_color[3],lw=1,label=r'$R=1,5\, A^{1/3}$')
plt.plot(x,radiosaturacion*x,'--',lw=1,c=jpac_color[4],label=r'$R=r_s\, A^{1/3}$')
x = np.linspace(1.5,343,100)
plt.plot(x**(1/3),radii(0.9071,1.105,-0.548,x),'-',lw=1,c=jpac_color[10],label=r'Radio de carga, $R=A^{1/3}(r_1+r_2/A^{2/3}+r_3/A^{4/3})$')
plt.scatter(dfrne['A']**(1/3),dfrne['radius'],marker='o',s=5,c=jpac_color[0],label='Radio de carga')
plt.scatter(dfres['A']**(1/3),dfres['radius'],marker='o',s=5,c=jpac_color[0])
plt.scatter(dfbw['A']**(1/3),dfbw['R'],marker='o',s=5,c=jpac_color[1],label='Radio de materia, Blatt y Weisskopf (1952)')
plt.scatter(dfnds['AD']**(1/3),dfnds['radio'],marker='o',s=5,c=jpac_color[3],label='Radio de materia, NDS167 (2020) 1')
plt.legend(loc='upper left',ncol=1,frameon=False)
plt.show()
#fig.savefig("RadioVsRadio.pdf", bbox_inches='tight')
```

Deformación nuclear

```python
def nucleodeformado(opcion):
    fig = plt.figure(figsize=plt.figaspect(1))  # Square figure
    ax = fig.add_subplot(111, projection='3d')
    # Coeficientes en a x**2 + b y**2 + c z**2 = 1 
    if opcion=='e':
        coefs = (1, 1, 1)  # esferico
        color = 0
    elif opcion=='p':
        coefs = (1, 1, 1/5)  # prolato
        color = 1
    else: 
        coefs = (1/2, 1/2, 1)  # oblato
        color = 3
        
    rx, ry, rz = 1/np.sqrt(coefs)
    u = np.linspace(0, 2 * np.pi, 100)
    v = np.linspace(0, np.pi, 100)
    x = rx * np.outer(np.cos(u), np.sin(v))
    y = ry * np.outer(np.sin(u), np.sin(v))
    z = rz * np.outer(np.ones_like(u), np.cos(v))
    ax.plot_surface(x, y, z,  rstride=4, cstride=4, color=jpac_color[color])
    max_radius = max(rx, ry, rz)
    for axis in 'xyz':
        getattr(ax, 'set_{}lim'.format(axis))((-max_radius, max_radius))

    plt.show()
    #fig.savefig('forma_nuclear.pdf', bbox_inches='tight') 

#   Selección entre: e = esferico; p = prolato; o = oblato
opcion = ['e','p','o']
nucleodeformado(opcion[0]); nucleodeformado(opcion[1]); nucleodeformado(opcion[2])
```
